# *TASK 1*

```python
url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
y1=pd.read_csv(url)
y1
```

```python
print(y1.isnull().sum())
```

```python
# print(y1.head())  this both above and below fn are used to cehck the columns name for smooth removal of empty data.
```

```python
# print(y1.columns.tolist())
```

```python
y1_clean=y1.drop(['Lat','Long'],axis=1)
y1_clean
```

```python
y1_clean['Province/State']=y1_clean['Province/State'].fillna(y1_clean['Country/Region'])
y1_clean
```

```python
y1_clean=y1_clean.ffill(axis=1)
y1_clean
```

```python
y1_clean['Total Cases']=y1_clean.iloc[:,4:].sum(axis=1)

pivot_country=pd.pivot_table(
    y1_clean,
    index='Country/Region',
    values='Total Cases',
    aggfunc='sum'
).sort_values('Total Cases',ascending=True)

print(pivot_country.head)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot top 10 countries
top_10 = pivot_country.head(10)
sns.barplot(x=top_10['Total Cases'], y=top_10.index)
plt.title("Top 10 Countries by COVID-19 Cases")
plt.show()
```

# *TASK 2*

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
y1=pd.read_csv(url)
y1_long = y1.melt(
    id_vars=['Country/Region', 'Province/State', 'Lat', 'Long'], 
    var_name='Date', 
    value_name='Cases'
)
y1_long['Date'] = pd.to_datetime(y1_long['Date'])
y1_country = y1_long.groupby(['Country/Region', 'Date'])['Cases'].sum().reset_index()


countries = ['US', 'India', 'Brazil', 'Germany', 'China', 'Japan','Tuvalu','Antarctica','Holy See','Nauru'] 
plt.figure(figsize=(14, 7))
sns.lineplot(data=y1_country[y1_country['Country/Region'].isin(countries)], 
             x='Date', y='Cases', hue='Country/Region')
plt.yscale('log')  # Remove if linear preferred
plt.title('Daily COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
y11_country = y1.drop(columns=['Province/State', 'Lat', 'Long']).groupby('Country/Region').sum()
y1_dates = y11_country.T
y1_dates.index = pd.to_datetime(y1_dates.index)
y1_monthly = y1_dates.resample('ME').sum().T
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(
    y1_monthly,
    cmap='YlOrRd',
    annot=False,  # Set to True to show values (may be crowded)
    fmt='g',
    linewidths=0.5
)
plt.title('Monthly COVID-19 Cases by Country')
plt.xlabel('Month')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
