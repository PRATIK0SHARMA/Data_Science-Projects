# *TASK 1*

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path = r"C:\Users\sharm\Documents\jupyter_python\PROJECTS\Sample - Superstore.csv"
try:
    df = pd.read_csv(path, encoding='utf-8')
except:
    df = pd.read_csv(path, encoding='latin1')
df
```

```python
print(df.head())
```

```python
top_selling=df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head()
top_selling
```

```python
top_selling.plot(kind='barh',figsize=(12,6),color='orange')
plt.title("TOP-5 SELLING PRODUCTS IN USA BY QUANTITY")
plt.xlabel('Total Units Sold')
plt.ylabel('Product Name')
plt.grid()
plt.show()
```

```python
df_top=df.head(500)
plt.figure(figsize=(12,6))
sns.scatterplot(x="Sales",y="Profit",hue="Quantity",data=df_top,legend=True,palette="Accent")
plt.title("SALES V/S PROFITS OF TOP (500) ROWS")
plt.grid()
plt.show()
```

# *TASK-3*

```python
import numpy as np
df['Profit Margin %']=(df['Profit']/df['Sales'])*100
df['Profit Margin %']=df['Profit Margin %'].replace([-np.inf,np.inf],np.nan).fillna(0)
```

```python
highest_prft_mrgn_pdct=df.loc[df['Profit Margin %'].idxmax()]
highest_prft_mrgn_pdct[['Product Name','Profit Margin %']]
```

```python
top_prdct=df.sort_values('Profit Margin %',ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Profit Margin %',y='Product Name',hue="Sales",data=top_prdct)
plt.title("TOP-10 PRODUCTS BY PROFIT MARGINS(%)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Category', y='Profit Margin %',data=df)
plt.title('PROFIT MARGIN BY CATEGORY')
plt.xticks(rotation=45)
plt.show()
```
